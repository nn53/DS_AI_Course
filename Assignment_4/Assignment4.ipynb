```python
# -------------------------------
# Assignment 4: Statistics & Probability
# -------------------------------

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load cleaned dataset
data = pd.read_csv("creditcard_cleaned.csv") 
data.head()
```

```python
print("\n=== Basic Statistics ===")

# Mean
mean_values = data.mean(numeric_only=True)
print("\nMean:\n", mean_values)
```

```python
# Median
median_values = data.median(numeric_only=True)
print("\nMedian:\n", median_values)
```

```python
# Mode
mode_values = data.mode(numeric_only=True).iloc[0]
print("\nMode:\n", mode_values)
```

```python
# Variance
variance_values = data.var(numeric_only=True)
print("\nVariance:\n", variance_values)
```

```python
# Calculate correlation of all features with target variable 'Class'
corr_matrix = data.corr()

# Sort correlation values (absolute) to find most related features
corr_with_class = corr_matrix["Class"].abs().sort_values(ascending=False)
print(corr_with_class.head(10))
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, cmap='coolwarm', center=0)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
top_features = corr_with_class[1:4].index  # Skip 'Class' itself
print("Top 3 correlated features:", top_features.tolist())

# Plot distribution for each
for feature in top_features:
    plt.figure(figsize=(6,4))
    sns.boxplot(x='Class', y=feature, data=data)
    plt.title(f'{feature} vs Class')
    plt.show()
```
