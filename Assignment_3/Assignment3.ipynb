```python
# -------------------------------
# Assignment 3: Data Visualization
# -------------------------------

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned dataset 
data = pd.read_csv("creditcard_cleaned.csv") 
data.head()
```

```python
sns.countplot(x='Class', data=data)
plt.title('Fraud vs Non-Fraud Transaction Count')
plt.xlabel('Transaction Type (0 = Non-Fraud, 1 = Fraud)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data['Amount'], bins=50, kde=True)
plt.title('Distribution of Transaction Amounts')
plt.xlabel('Transaction Amount ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(data.corr(), cmap='coolwarm', center=0)
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(x='Class', y='Amount', data=data)
plt.title('Transaction Amounts by Class')
plt.xlabel('Class (0=Non-Fraud, 1=Fraud)')
plt.ylabel('Amount ($)')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='Time', y='Amount', hue='Class', data=data, alpha=0.6)
plt.title('Transaction Amounts Over Time')
plt.xlabel('Time (seconds)')
plt.ylabel('Amount ($)')
plt.legend(title='Class')
plt.show()
```
